```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_mood_smartphone.csv")
```

```python
# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])
```

```python
df
```

```python
pivot_df = df.pivot_table(index=['id', 'date'], columns='variable', values='value', aggfunc='sum').reset_index()
aggregations = {
    'mood': 'mean',
    'circumplex.arousal': 'mean',
    'circumplex.valence': 'mean',
    'activity': 'mean',
    'screen': 'sum',
    'call': 'sum',
    'sms': 'sum',
    'appCat.builtin': 'sum',
    'appCat.communication': 'sum',
    'appCat.entertainment': 'sum',
    'appCat.finance': 'sum',
    'appCat.game': 'sum',
    'appCat.office': 'sum',
    'appCat.other': 'sum',
    'appCat.social': 'sum',
    'appCat.travel': 'sum',
    'appCat.unknown': 'sum',
    'appCat.utilities': 'sum',
    'appCat.weather': 'sum'
    }

    # Group by 'ID' and 'Date', and apply the aggregation functions
daily_aggregated_df = pivot_df.groupby(['id', 'date']).agg(aggregations).reset_index()

    # Drop rows where 'mood' is NaN
daily_aggregates = daily_aggregated_df.dropna(subset=['mood'])
daily_aggregates 
```

```python
# Assuming 'df' is your DataFrame and 'variable' is a column in your DataFrame
score_vars = ['mood', 'circumplex.arousal', 'circumplex.valence']

# Filtering DataFrame for score variables
df_scores = df[df['variable'].isin(score_vars)]

# Filtering DataFrame for non-score variables
df_non_scores = df[~df['variable'].isin(score_vars)]

# Pivot with aggregation for scores
pivot_scores = df_scores.pivot_table(index=['id', 'date'], columns='variable', values='value', aggfunc='mean').reset_index()

# Pivot with aggregation for non-scores
pivot_non_scores = df_non_scores.pivot_table(index=['id', 'date'], columns='variable', values='value', aggfunc='sum').reset_index()

# Merge the two pivoted DataFrames on 'id' and 'date'
df_final = pd.merge(pivot_scores, pivot_non_scores, on=['id', 'date'], how='outer')

df_final
```
